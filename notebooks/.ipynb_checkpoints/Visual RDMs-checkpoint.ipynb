```python
def plot_roi(func_path, roi_path):
    func = nl.image.load_img(func_path)
    mask = nl.image.load_img(roi_path)
    func_data = func.get_fdata()
    mask_data = mask.get_fdata()
    roi = func_data[mask.get_fdata() == 1]
    px.line(roi.T)
```

```python
import re

with open('/galitylab/data/studyforrest-movie/forrest_gump_script.txt', 'r') as f:
    script = f.read().replace('\t', '')
#     for i in range(50):
    script = script.replace('\n'+15*' ','\n')
    
    dataset = {
        'character': [],
        'speech': []
    }
    
    for match in re.finditer(r'(\n +[A-Z (.)]+\n)', script):
        cls = match.group(0)
        cls = cls.replace('\n', '').replace(22*' ', '')
        
        speech_start_idx = match.end(0)
        speech_end_idx = script.find('\n\n', speech_start_idx)
        speech = script[speech_start_idx : speech_end_idx]
        speech = speech.replace('\n', '').replace(10*' ', '')
        
        dataset['character'].append(cls)
        dataset['speech'].append(speech)
        
dataset = pd.DataFrame(dataset)
dataset
dataset.to_csv('/galitylab/data/studyforrest-movie/forrest_gump_characters_speech.csv')
```
